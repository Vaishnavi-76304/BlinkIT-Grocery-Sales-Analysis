```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Blinkit = pd.read_excel('/content/BlinkIT Grocery Data.xlsx')
Blinkit.head()
```

```python
Blinkit.shape
```

```python
Blinkit = Blinkit.drop_duplicates()
Blinkit.head()
```

```python
Blinkit = Blinkit.dropna()
Blinkit.head()
```

```python
Blinkit['Item Weight'].fillna(Blinkit['Item Weight'].mean(), inplace=True)
```

```python
Blinkit.dropna(subset=['Item Weight'], inplace=True)
```

```python
Blinkit['Item Fat Content'] = Blinkit['Item Fat Content'].replace({'LF':'Low Fat','reg':'Regular','low fat' : 'Low Fat'})
Blinkit['Item Fat Content'].head()
```

```python
Blinkit['Rating'] =Blinkit['Rating'].astype(int)
```

```python
Blinkit.nunique()
```

```python
#Top Rating Based on Item Type
Top_rating = Blinkit.groupby('Item Type')['Rating'].mean().sort_values(ascending= False)
Top_rating.head()
```

```python
#Top 10 Sales record
Top_Sales = Blinkit['Sales'].sort_values(ascending=False)
Top_Sales.head(10)
```

```python
Total_Sales_by_Item_Type = Blinkit.groupby('Item Type')['Sales'].mean().sort_values(ascending = True)
Total_Sales_by_Item_Type.head()
```

```python
Item_weight_by_Item_type = Blinkit.groupby('Item Weight')['Item Type'].max().sort_values(ascending = True)
Item_weight_by_Item_type.head()
```

```python
Outlet_size = Blinkit.groupby('Outlet Size')['Item Fat Content'].max().sort_values(ascending = True)
Outlet_size
```

```python
Blinkit.head()
```

```python
#Data Visualization
Blinkit['Item Fat Content'].value_counts().plot(kind='bar')
```

```python
#Blinkit['Outlet Size'].value_counts().plot(kind= 'bar')
sns.countplot(x = 'Outlet Size', data= Blinkit,palette= 'viridis',order=Blinkit['Outlet Size'].value_counts().index)
plt.title('Distribution of outlet Type')
```

```python
#Blinkit['Outlet Type'].value_counts().plot(kind= 'bar')
sns.countplot(x= 'Outlet Type',data = Blinkit,palette= 'viridis',order= Blinkit['Outlet Type'].value_counts().index)
plt.title('Distribution of Outlet Type')
```

```python
sns.countplot(y = 'Item Type', data = Blinkit,palette='viridis', order=Blinkit['Item Type'].value_counts().index)
plt.title('Distribution of Item Type')
```

```python
grouped = Blinkit.groupby('Item Type')['Sales'].sum()
sorted_grouped = grouped.sort_values(ascending=False)
top_10 = sorted_grouped.head(10)
plt.figure(figsize=(12,6))

sns.barplot(x=top_10.index, y=top_10.values, palette='viridis')
plt.xticks(rotation = 45)
plt.title('Top 5 Item by Total Sales')
plt.ylabel('Total_Sales')
plt.xlabel('Item Type')

```

```python
sns.boxplot(x = 'Outlet Type',y = 'Sales',data= Blinkit,palette='viridis')
plt.xticks(rotation=45)
plt.title('Sales Distribution By Outlet Type')
```

```python
sns.scatterplot(x = 'Item Visibility', y = 'Sales',palette='viridis',data = Blinkit,alpha = 0.5)
plt.title('Item Visibility vs Sales')
```

```python
sns.barplot(x = 'Outlet Location Type',y = 'Sales',palette= 'viridis',data= Blinkit,estimator= 'mean')
plt.title('Outlet Location Type By Avg Sales')
```

```python
sns.violinplot(x= 'Outlet Size',y= 'Rating',palette= 'viridis', data = Blinkit)
plt.title('Outlet Size By Rating')
```

```python
sns.histplot(Blinkit['Item Weight'].dropna(), bins=30,palette= 'viridis',kde=True)
plt.title('Distribution of Item Weight')
```
